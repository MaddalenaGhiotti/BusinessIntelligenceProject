missing value:
-  smoking_history ---> no Info : cambiare con Nan
- Insulin_Sensitivity_Est --> NaN

```python
dataSetTrain.replace(to_replace = 'No Info', value = np.nan, inplace = None)
dataSetTrain.isnull().sum() 
```

```python

attributes = dataSetTrain.columns

# Create scatter/bubble plots for each attribute against 'Age'
for attribute in attributes:
    if attribute != 'Age':  # Skip the 'Age' column to avoid plotting Age vs Age
        plt.figure(figsize=(10, 6))
        plt.scatter(dataSetTrain[attribute], dataSetTrain['age'], s=50, alpha=0.5, label=attribute)
        plt.xlabel(attribute, fontsize=10)
        plt.ylabel('Age', fontsize=10)
        plt.title(f'Scatter/Bubble Plot with {attribute} on X-axis and Age on Y-axis')
        plt.legend()
        plt.show()
 # Fix the 'Age' attribute on the y-axis
```

```python
attributes2= dataSetTrain.columns

# Create scatter/bubble plots for each attributE
for attribute2 in attributes2:
    plt.figure(figsize=(10, 6))
    plt.bar(dataSetTrain.index, dataSetTrain[attribute2], alpha=0.7)  # 'alpha' per la trasparenza delle barre
    plt.xlabel('Utente (indice)')
    plt.ylabel(attribute2)
    plt.title(f'Grafico a barre di {attribute2}')
    plt.show()
```
